```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.preprocessing import PowerTransformer
```

```python
from sklearn.preprocessing import QuantileTransformer
```

```python
np.random.seed(0)

df = np.random.exponential(size= 1000)

df = pd.DataFrame(df , columns = ['values'])

df.shape
```

```python
sns.histplot( df , kde= True)
```

```python
pt_box_cox = PowerTransformer( method= 'box-cox') # Only handle positive values.

pt_yeo_johnson = PowerTransformer( method='yeo-johnson')

qt_transformer = QuantileTransformer( output_distribution= 'normal')
```

```python
df['box-cox'] = pt_box_cox.fit_transform(df[['values']] + 1) # It is necessary to add 1 with the box-cox transformation.

df['yeo_johnson'] = pt_yeo_johnson.fit_transform(df[['values']])

df['Quantile'] = qt_transformer.fit_transform(df[['values']])
```

```python
df.head()
```

```python
for i in df.columns:
    sns.histplot( df[i] , kde= True)
    plt.show()

```
